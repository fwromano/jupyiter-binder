```python

def fpi(g,x0, iters, decplaces):
    """
    computes and prints k iterations of g starting at x0

    Args:
        g: the function to be iterated
        x0: initial value of x
        iters: number of iterations
        decplaces: number of decimal digits to display
    Return:
        final value of x after iterations
    """

    formatstr = '{:3d}{: '+ str(decplaces+7) + '.' + str(decplaces) + 'f}'

    x=x0;
    print(formatstr.format(0,x))
    for i in range(1, iters+1):
        x = g(x)
        print(formatstr.format(i,x))
    return x
```

```python
def g(x): return 1/(x**4+1) # 2a
fpi(g, .75, 1000, 8) 
```

```python
import math
def g(x): return (math.sin(x)-5)/(5) #2b
fpi(g, .5, 10, 8)
```

```python
import math
def g(x): return (3-math.log(x)/x) #2c
fpi(g, 1, 10, 8) 
```

```python
def g(x): return (2*x + 2/x**2)/3 # 4a
fpi(g, 1, 10, 8)
```

```python
def g(x): return (2*x + 3/x**2)/3 #4b
fpi(g, 1, 10, 8)
```

```python
def g(x): return (2*x + 5/x**2)/3 #4c
fpi(g, 1, 10, 8)
```

```python
def g(x): return (2*x**3-1)/6 #6a1
fpi(g, 1.5, 10, 8)
```

```python
def g(x): return (6*x+1)/(2*x**2) #6a2
fpi(g, -1, 200, 8)
```

```python
def g(x): return 1/(2*x**2-6) #6a3
fpi(g, 2, 200, 8)
```

```python
def g(x): return math.e**(x-2)+x**3 #6b1
fpi(g, -1, 200, 8)
```
